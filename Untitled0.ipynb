```python
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Load dataset
url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/flights.csv"
data = pd.read_csv(url)

# Show first rows
print(data.head())

# Create month number
data['Month_Number'] = np.arange(len(data))

# Features and target
X = data[['Month_Number']]
y = data['passengers']

# Train model
model = LinearRegression()
model.fit(X, y)

# Predict future 12 months
future_months = np.arange(len(data), len(data)+12).reshape(-1,1)
future_predictions = model.predict(future_months)

# Print predictions
print("\nFuture Predictions:")
print(future_predictions)

# Plot actual data
plt.figure(figsize=(10,5))
plt.plot(data['Month_Number'], y, label='Actual Data')

# Plot future predictions
plt.plot(future_months, future_predictions, label='Predicted Data')

plt.xlabel("Month Number")
plt.ylabel("Passengers")
plt.title("Predictive Analytics Using Historical Data")
plt.legend()
plt.show()
```
